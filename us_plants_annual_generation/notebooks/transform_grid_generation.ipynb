```python
import pandas as pd
```

```python
df_plants_anual_gen = pd.read_excel('eGRID2021_data.xlsx', sheet_name='GEN21')
```

```python
df_plants_anual_gen_v1 = df_plants_anual_gen[['Plant name', 'Generator annual net generation (MWh)']]
df_plants_anual_gen_v1 = df_plants_anual_gen_v1.loc[1:].reset_index(drop = True)
```

```python
df_plants_anual_gen_v1['Generator annual net generation (MWh)'] = \
df_plants_anual_gen_v1['Generator annual net generation (MWh)'].fillna(0)

df_plants_anual_gen_v1['Generator annual net generation (MWh)'] = \
df_plants_anual_gen_v1['Generator annual net generation (MWh)'].astype('int')
```

```python
df_plants_anual_gen_v1.groupby(['Plant name']).agg({'Generator annual net generation (MWh)':'sum'}).reset_index().sort_values(by = 'Generator annual net generation (MWh)', ascending=False).reset_index(drop = True).loc[]
```
